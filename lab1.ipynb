```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline

dist = np.array([5, 8, 15, 20, 24])
beam_diameter = np.array([.6, .45, .2, .35, .65])
beam_radius = beam_diameter / 2

# Create smooth x values
dist_smooth = np.linspace(dist.min(), dist.max(), 300)
# Fit spline
spline = make_interp_spline(dist, beam_radius, k=2)  # k=2 for quadratic (parabola-like)
beam_radius_smooth = spline(dist_smooth)

plt.figure(figsize=(10, 5))
plt.plot(dist_smooth, beam_radius_smooth, color='b', label='Smooth Beam Radius Curve')
plt.scatter(dist, beam_radius, color='red', label='Original Points')
plt.title('Beam Profile (Smooth Curve)')
plt.xlabel('Distance (cm)')
plt.ylabel('Beam Radius (cm)')
plt.legend()
plt.savefig('Beam Profile Smooth Curve.png')  # Save the figure as a PNG file
plt.show()
```
